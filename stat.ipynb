```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skew, kurtosis

# analytic values
mu_true = 0
var_true = 1
skew_true = 0
kurt_true = 0  # excess kurtosis

Ks = [4**n for n in range(1, 11)]

means = []
vars_ = []
skews = []
kurts = []

np.random.seed(0)

for K in Ks:
    x = np.random.normal(0, 1, K)
    means.append(np.mean(x))
    vars_.append(np.var(x, ddof=0))   # population variance
    skews.append(skew(x, bias=True))
    kurts.append(kurtosis(x, bias=True))  # excess kurtosis

invK = 1 / np.array(Ks)

# plotting
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

axs[0, 0].plot(invK, means, 'o-', label='Sample')
axs[0, 0].axhline(mu_true, color='r', linestyle='--', label='Analytic')
axs[0, 0].set_title('Mean')
axs[0, 0].set_xlabel('1/K')

axs[0, 1].plot(invK, vars_, 'o-', label='Sample')
axs[0, 1].axhline(var_true, color='r', linestyle='--', label='Analytic')
axs[0, 1].set_title('Variance')
axs[0, 1].set_xlabel('1/K')

axs[1, 0].plot(invK, skews, 'o-', label='Sample')
axs[1, 0].axhline(skew_true, color='r', linestyle='--', label='Analytic')
axs[1, 0].set_title('Skewness')
axs[1, 0].set_xlabel('1/K')

axs[1, 1].plot(invK, kurts, 'o-', label='Sample')
axs[1, 1].axhline(kurt_true, color='r', linestyle='--', label='Analytic')
axs[1, 1].set_title('Excess Kurtosis')
axs[1, 1].set_xlabel('1/K')

for ax in axs.flat:
    ax.legend()
    ax.grid(True)

plt.tight_layout()
plt.show()
```
